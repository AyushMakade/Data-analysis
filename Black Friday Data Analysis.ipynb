## Gender Ratio Pie Chart

```python
df.groupby('Gender').size().plot(kind = 'pie', autopct = '%.1f%%', title = 'Gender Ratio', figsize = (6,6))
```

```python
df_pr = df.groupby('Gender').mean()['Purchase']
plt.style.use('ggplot')
plt.pie(df_pr, labels = ['Female', 'Male'], colors = ['g','c'], autopct='%.1f%%', shadow=True)
plt.axis('equal')
plt.legend()
plt.title('Gender Ratio')
plt.show()
```

## Gender Ratio Bar Graph

```python
df.groupby('Gender').size().plot(kind = 'bar', figsize = (6,6))
```

## Relation between gender and purchase

```python
df.groupby('Gender').sum()['Purchase'].plot(kind = 'pie', autopct = "%0.1f%%")
```

## Relation between gender and purchase on average

```python
df.groupby('Gender').mean()['Purchase']
```

```python
df.groupby('Gender').mean()['Purchase'].plot(kind = 'pie', autopct = "%0.1f%%")
```

## Purchase Distribution by Age

```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (12, 6), title = 'Purchase Distribution by Age')
```

## Unique products purchased by different age group

```python
lst = []
for i in df['Age'].unique():
    lst.append([i, df[df['Age'] == i]['Product_ID'].nunique()])
    
data = pd.DataFrame(lst , columns = ['Age','Products'])
data.plot.bar(x = 'Age', figsize = (8,6))
```

## Total amount spent by different age group

```python
df.groupby('Age').sum()['Purchase'].plot(kind = 'bar', figsize = (12, 6), title = 'Amount Spend by Age')
```

## Mean amount spent by different age group - Bar Graph

```python
df.groupby('Age').mean()['Purchase'].plot(kind = 'bar', figsize = (12, 6), title = 'Amount Spend by Age')
```

## Mean amount spent by different age group - Pie Chart

```python
df.groupby('Age').mean()['Purchase'].plot(kind = 'pie', autopct = '%0.1f%%')
```
